```python
import sys, os
import datetime
from datetime import date
sys.path.append(os.environ['minotaur'])

import yaml

%matplotlib inline
from matplotlib import pyplot as plt

import pandas as pd
import numpy as np

from pylab import rcParams
rcParams['figure.figsize'] = 15,5

from matplotlib.dates import DAILY
from datetime import datetime

from dbs import redshift
redshift.connect()
```

```python
df = redshift.execute("""
select id,partner_id,email from gravity.partner_contacts where email is not null
""")
```

```python
df['email'] = df['email'].str.lower()
df.head()
```

```python
df['duplicate']= df.duplicated(subset=['partner_id','email'],keep=False)
```

```python
dp = df[df.duplicate == True]
```

```python
dp.sort_values('email')
```

```python
dp.to_csv('duplicate_contacts.csv')
```
