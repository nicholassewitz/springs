```python
import sys, os
import datetime
from datetime import date
sys.path.append(os.environ['minotaur'])

import yaml

%matplotlib inline
from matplotlib import pyplot as plt

import pandas as pd
import numpy as np
pd.options.display.max_rows = 999

from pylab import rcParams
rcParams['figure.figsize'] = 15,5

from matplotlib.dates import DAILY
from datetime import datetime

from dbs import redshift
redshift.connect()
```

```python
q = redshift.execute("""
select * from marketo.clicks_email_link""")
q1 = redshift.execute("""
select * from marketo.email_bounces""")
q2 = redshift.execute("""
select * from marketo.email_delivers""")
q3 = redshift.execute("""
select * from marketo.email_opens""")
q4 = redshift.execute("""
select * from marketo.email_sends""")
q5 = redshift.execute("""
select * from marketo.email_unsubscribes""")
q6 = redshift.execute("""
select * from marketo.form_submissions_delete""")
```

```python
q = q.replace('',np.nan)
q = q.fillna(np.nan)
q = q.set_index(['id'],drop=True)

q1 = q1.replace('',np.nan)
q1 = q1.fillna(np.nan)
q1 = q1.set_index(['id'],drop=True)

q2 = q2.replace('',np.nan)
q2 = q2.fillna(np.nan)
q2 = q2.set_index(['id'],drop=True)

q3 = q3.replace('',np.nan)
q3 = q3.fillna(np.nan)
q3 = q3.set_index(['id'],drop=True)

q4 = q4.replace('',np.nan)
q4 = q4.fillna(np.nan)
q4 = q4.set_index(['id'],drop=True)

q5 = q5.replace('',np.nan)
q5 = q5.fillna(np.nan)
q5 = q5.set_index(['id'],drop=True)

q6 = q6.replace('',np.nan)
q6 = q6.fillna(np.nan)
q6 = q6.set_index(['id'],drop=True)
```

```python
q.to_csv('clicks_email_link_no_na.csv')


q1.to_csv('email_bounces_no_na.csv')


q2.to_csv('email_delivers_no_na.csv')


q3.to_csv('email_opens_no_na.csv')


q4.to_csv('email_sends_no_na.csv')


q5.to_csv('email_unsubscribes_no_na.csv')


q6.to_csv('form_submissions_no_na.csv')
```
